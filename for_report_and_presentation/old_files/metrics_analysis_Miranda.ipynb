```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_metrics(csv_file_path, save_plot = False, output_file_path = None):
    # Read the CSV file
    data = pd.read_csv(csv_file_path, sep='\t')  # Use tab separator for your data

    epoch = data['epoch']
    training_loss = data['training_loss']
    metabolite_coverage_score = data['metabolite_coverage_score']
    top_1_metabolite_score = data['top_1_metabolite_score']
    validation_loss = data['validation_loss']

    # Plot the training and validation loss against epochs
    plt.figure(figsize=(5, 3))
    
    # plt.plot(epoch, training_loss, label='Training Loss', color='red')
    plt.plot(epoch, metabolite_coverage_score, label='Metabolite Coverage', color='blue')
    plt.plot(epoch, top_1_metabolite_score, label='Top-1 Metabolite', color='green')
    plt.plot(epoch, validation_loss, label='Validation Loss', color='orange')
    
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss vs. Epochs')
    plt.legend()
    plt.grid(True)
    plt.ylim(0, 1)
    # plt.xlim(0,30)
    
    # Show the plot
    plt.show()

    if save_plot:
        # Save the figure
        plt.savefig(output_file_path, format='png')
```

```python
import sys
import os

def plot_metrics_report(csv_file_path, save_figure):
    # Read the CSV file
    data = pd.read_csv(csv_file_path, sep='\t')  # Use tab separator for your data

    plt.rcParams.update({"font.size": 12, "font.family": "serif", "font.serif": ["NewComputerModern10"]})
    # plt.rcParams.update({'font.size': 12, 'font.family': 'sans-serif'})

    plt.figure(figsize=(6, 3.5))

    colour_scheme = ['#52B297', '#8E4A93', '#009EE7', '#EF766E']

    epoch = data['epoch']
    training_loss = data['training_loss']
    metabolite_coverage_score = data['metabolite_coverage_score']
    validation_loss = data['validation_loss']

    # Plot the training and validation loss against epochs
    plt.figure(figsize=(5, 3))
    
    # plt.plot(epoch, training_loss, label='Training Loss', color='red')
    plt.plot(epoch, metabolite_coverage_score, label='Metabolite Coverage', color=colour_scheme[3])
    plt.plot(epoch, validation_loss, label='Validation Loss', color=colour_scheme[2])
    
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True)
    plt.ylim(0, 1)
    # plt.xlim(0,30)

    if save_figure is not None:
        os.makedirs('figures', exist_ok=True)
        plt.savefig(f'figures/{save_figure}.eps', format='eps')

    # Show the plot
    plt.show()
```
