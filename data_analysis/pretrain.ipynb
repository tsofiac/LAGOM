```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_and_save_loss(csv_file_path, save_plot = False, output_file_path = None):
    # Read the CSV file
    data = pd.read_csv(csv_file_path, sep='\t')  # Use tab separator for your data

    epoch = data['epoch']
    training_loss = data['training_loss']
    validation_loss = data['validation_loss']

    # Plot the training and validation loss against epochs
    plt.figure(figsize=(5, 3))
    
    plt.plot(epoch, training_loss, label='Training Loss', color='blue')
    plt.plot(epoch, validation_loss, label='Validation Loss', color='orange')
    
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss vs. Epochs')
    plt.legend()
    plt.grid(True)
    plt.ylim(0, 1)
    #plt.xlim(0, 20)
    
    # Show the plot
    plt.show()

    if save_plot:
        # Save the figure
        plt.savefig(output_file_path, format='png')
```

```python
csv_file_path = '/projects/cc/se_users/carlsson_ksmq649/MasterThesis/results/pretraining/base/forward_prediction/version_9/logged_train_metrics.csv'
plot_and_save_loss(csv_file_path)
#ChemVA
```
