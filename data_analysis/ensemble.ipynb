```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_identified_metabolites_ensemble(nr_of_metabolites):

    x = np.arange(1, len(nr_of_metabolites)+1)

    plt.plot(x, nr_of_metabolites, 'o-', color='mediumseagreen')
    plt.ylabel('Number of Identified Metabolites')
    plt.xlabel('Number of Models')
    plt.ylim(0, 271)
    plt.xticks(np.arange(1, len(nr_of_metabolites)+1))
    plt.grid()
    plt.show()
```

```python
nr_of_metabolites = [120, 134, 143, 145, 147]
plot_identified_metabolites_ensemble(nr_of_metabolites)
```
