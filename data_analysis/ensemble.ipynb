```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_identified_metabolites_ensemble(nr_metabolites_each, nr_metabolites_ensamble):

    x = np.arange(1, len(nr_metabolites_each)+1)

    plt.plot(x, nr_metabolites_each, 'o-', color='slateblue', label='Model')
    plt.plot(x, nr_metabolites_ensamble, 'o-', color='mediumseagreen', label='Ensemble')
    plt.ylabel('Number of Identified Metabolites')
    plt.xlabel('Number of Models')
    plt.ylim(0, 271)
    plt.xticks(np.arange(1, len(nr_metabolites_ensamble)+1))
    plt.grid()
    plt.legend()
    plt.show()
```

```python
# baseline - aug 0.5 - mask 0.1 - PG - PP - csp3 - logP
nr_metabolites_each = [120, 113, 114, 116, 117, 107, 111]
nr_metabolites_ensemble = [120, 134, 143, 145, 147, 148, 151]
plot_identified_metabolites_ensemble(nr_metabolites_each, nr_metabolites_ensemble)
```
